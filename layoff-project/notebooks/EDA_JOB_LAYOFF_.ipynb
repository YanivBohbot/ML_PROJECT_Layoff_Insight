```python
# -------------------------
# STEP 1 — Clean percentage column
# -------------------------

# Convert "25%" → 25.0
def clean_percentage(val):
    if pd.isna(val):
        return np.nan
    return float(str(val).replace('%',''))

df["perc_laid_off"] = df["percentage_laid_off"].apply(clean_percentage)

# -------------------------
# STEP 2 — Handle missing percentages
# -------------------------

# If both % and total_laid_off are missing → assume no layoffs → set 0%
df.loc[
    (df["perc_laid_off"].isna()) & (df["total_laid_off"].isna()),
    "perc_laid_off"
] = 0.0

# -------------------------
# STEP 3 — Create layoff severity class
# -------------------------

# Define severity mapping:
# 0 → Low (≤10%)
# 1 → Medium (10–50%)
# 2 → High (>50%)
# 3 → Unknown (missing % but layoffs occurred)

def severity_class(p, total):
    if pd.isna(p):
        return 3    # Unknown severity
    elif p <= 10:
        return 0    # Low severity
    elif p <= 50:
        return 1    # Medium severity
    else:
        return 2    # High severity

df["layoff_severity"] = df.apply(
    lambda row: severity_class(row["perc_laid_off"], row["total_laid_off"]),
    axis=1
)

# -------------------------
# STEP 4 — Check class distribution
# -------------------------

print(df["layoff_severity"].value_counts())
print(df["layoff_severity"].value_counts(normalize=True) * 100)

# Optional: preview new columns
print(df[["percentage_laid_off", "perc_laid_off", "layoff_severity"]].head(20))
```

```python
# How many missing % rows actually had layoffs?
missing = df[df["percentage_laid_off"].isna()]

print("→ Missing % rows with layoffs reported:", missing[~missing["total_laid_off"].isna()].shape[0])
print("→ Missing % rows with NO layoffs reported:", missing[missing["total_laid_off"].isna()].shape[0])
```

Visualize how layoffs evolved over  years.

```python

# Parse the date column
df["date"] = pd.to_datetime(df["date"])

# Group layoffs by year
layoffs_per_year = df.groupby(df["date"].dt.year).size()

print(layoffs_per_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
layoffs_per_year.plot(kind='bar')
plt.title("Number of Layoff Records per Year")
plt.ylabel("Number of Records")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```

Group Layoffs by Industry

```python
def map_industry_type(industry):
    if not isinstance(industry, str):
        return "Unknown"

    industry = industry.lower()

    if "travel" in industry or "hospitality" in industry:
        return "Travel"
    elif "sales" in industry or "retail" in industry:
        return "Sales"
    elif "security" in industry or "cyber" in industry:
        return "Security"
    elif "marketing" in industry or "advertising" in industry:
        return "Marketing"
    elif "logistic" in industry or "delivery" in industry:
        return "Logistics"
    elif "health" in industry or "med" in industry:
        return "Healthcare"
    elif "data" in industry or "analytics" in industry:
        return "Data"
    elif "ai" in industry or "ml" in industry or "artificial" in industry:
        return "AI"
    else:
        "Other"

```

```python
df["industry_type"] = df["industry"].apply(map_industry_type)
```

```python
industry_type_year = (
    df.groupby(["industry_type", "year"])["total_laid_off"]
    .sum()
    .reset_index()
)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=industry_type_year, x="year", y="total_laid_off", hue="industry_type", marker="o")
plt.title("Layoffs by Industry Type Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Layoffs")
plt.legend(title="Industry Type")
plt.tight_layout()
plt.show()
```
